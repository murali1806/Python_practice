```python
##list in python

playlist=["song1", "song2", "song3", "song4", "song5"]
food=["pizza", "burger", "pasta", "salad", "sushi"]
location=["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"]

##duplicate can be present in list, order will maintain like 0,1,2,3,4
#list is mutable, we can change the values in the list


a=[1,"string",1.5,True,False]## can define any type of data in the list

print("playlist:",playlist)
print("food:",food)
print("location:",location)
print ("a:",a)
```

```python
##list methods
##append method will add new value at the end of the list
playlist.append("songs6")
print("playlist after append:",playlist)
```

```python
#insert method will add the value at the specific index
playlist.insert(2, "songs7") # 2 is the index
print("playlist after insert:",playlist)
```

```python
##remove method will remove the value from the list
playlist.remove("song2")
print("playlist after remove:",playlist)
```

```python
##pop method will remove the value from the list based on index
playlist.pop(3) # 3 is the index
print("playlist after pop:",playlist)
```

```python
##reverse methood will reverse the order of the list
playlist.reverse()
print("playlist after reverse:",playlist)
##count
print("count",playlist.count("songs7"))
```

```python
##list slicing
print(playlist)
print("playlist first 2 songs:",playlist[0:2]) #[0:2] means 0 to 1 index value will be printed

#1 2 3 4 5 values in the list
#0 1 2 3 4 positive index 
#-4 -3 -2 -1 negative index
```

```python
uber={"chennai","chennai","bangalore","delhi","mumbai"}
print(set(uber)) ## set is define so duplicate is removed.
```

```python
uber1={"chennai","chennai","bangalore","delhi","mumbai"}
main_uber={"salem","delhi","HYD","mumbai"}

print("union:",uber1.union(main_uber)) ##union method will combine the two sets and remove the duplicate values
print("intersection:",uber1.intersection(main_uber)) ##intersection method will give the common values in the two sets
print("difference:",uber1.difference(main_uber)) ##difference method will give the values which is present in uber1 but not in main_uber
uber1.add("kerala")
print("add",uber1)
uber1.remove("delhi")
print("remove:",uber1)
##we can't fetch the value based on index in set because it is unordered collection of data type uber1[1] will give error
```

```python
##dictonary in python
##dictonary is unordered collection of data type that is iterable.
# mutable and has no duplicate elements
## it will have key and value pair, key should be unique and value can be duplicate
##example: dictonary={"key1":"value1","key2":"value2","key3":"value3"}
```

```python
travel={"trip_id":"A1",
        "pickup":"chennai",
        "drop":"airport",
        "fare":500,
        "driver":"John",
        "status":"completed"
        }

print(travel)
print(travel["driver"]) ##fetch the value based on key
print(travel.get("driver"))
print(travel.get("location"))###if we try to fetch the value based on key which is not present in the dictonary it will give None
print(travel.keys())
print(travel.values())
```

```python
for key,value in travel.items():##items method will give the key and value pair in the dictonary
    print(key,":",value)
```

```python
travel.update({"car":"Toyota"})##update method will work as upsert and add the new key and value pair in the dictonary
travel.update({"fare":600}) 
print (travel)
```

```python
travel.pop("status") ##pop method will remove the key and value pair based on key
print(travel)
```

```python
##duplicate check 
travel={"trip_id":"A1",
        "pickup":"chennai",
        "drop":"airport",
        "fare":500,
        "driver":"John",
        "status":"completed",
        "trip_id":"A2", ## duplicate key will be overwritten with the last value in output
        }

for k,v in travel.items():
    print(k,":",v)
```

```python
travel.update({"drop":["airport","central","metro"]}) ## we can add list as value in the dictonary
for k,v in travel.items():
    print(k,":",v)

##take only chennai from the list in drop key
print(travel["drop"][1])
```

```python
##can we iterate the list in the drop key
for i in travel["drop"]:
    print(i)
```

```python
##dictonary inside dictonary or we can give list as dictonary 

trips = {
    "UB001": {
        "trip_id": "UB001",
        "pickup": "Chennai",
        "drop": "Airport",
        "fare": 430
    },
    "UB002": {
        "trip_id": "UB002",
        "pickup": "Tambaram",
        "drop": "Central",
        "fare": 320
    },
    "UB003": {
        "trip_id": "UB003",
        "pickup": "T-Nagar",
        "drop": "Velachery",
        "fare": 210
    }
}

print(trips)
## UB002 what is fare?
print("UB002", trips["UB002"]["fare"]) ## need to give both the key
## id pickup from t-nagar --> drop in Velachery

for id,details in trips.items(): ## id is for out key and details for inner key pair
    print(f"{id} \npickup from {details["pickup"]} --> drop in {details["drop"]}")
```
