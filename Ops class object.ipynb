```python
class student:  ##class definition
    def say_hello(self): ##method of class, self is a reference to the current instance of the class
        print("Hello, I am a student.")

s1=student() #calling the class to create an object S1 of the class
s1.say_hello() #s1.say_hello(s1) ##calling the method of the class using the object S1


##procedure 
def team(): ## if defined outside the class, it is a function
    print("We are a team of students.")

team()
```

```python
##constructor is a special method that is called when an object is created. It is used to initialize the attributes of the class. 
#The __init__ method is called automatically when an object is created.
#if same attributes used in mulitiple methods use constructor
class student:
    def __init__(self, name, age, grade): ##constructor method, it is called when an object is created. It is used to initialize the attributes of the class.
        self.name = name
        self.age = age
        self.grade = grade

    def get_grade(self): ##method get the fields from above contructor
        print(f"{self.name} is in grade {self.grade}")

    def get_age(self): #methods
        print(f"{self.name} age is {self.age}")

#creating mulitple objects of the class student with different attributes
#create memory for each object and store the attributes in it
S1=student("murali",26, "A") #creating an object S1 of the class student
S2=student("sai", 25, "B") #creating an object S2 of the class student
S3=student("kumar", 27, "C") #creating an object S3 of the class student


S1.get_grade() #calling the method get_grade() of the class student using the object S1
S2.get_grade() #calling the method get_grade() of the class student using the object S2
S3.get_grade() #calling the method get_grade() of the class student using the object

S1.get_age() 
S2.get_age() 
S3.get_age() #calling the method get_age() of the class student using the object S1, S2, S3   
```

```python
##without constructor, we can create an object of the class student and then set the attributes using the methods of the class. But with constructor, we can set the attributes while creating the object of the class student.

class math:
    def add(self,a,b):
        return a+b
    def sub(self,a,b):
        return a-b
    def mul(self,a,b):
        return a*b
    def div(self,a,b):
        return a/b
calculator=math() #creating an object calculator of the class math
print("addition:",calculator.add(1,3)) #calling the method add() of the class math using the object calculator
print("subtraction:",calculator.sub(1,3)) #calling the method sub() of the class math using the object calculator
print("mulitply:",calculator.mul(1,3)) #calling the method mul() of the class math using the object calculator
print("division:",calculator.div(1,3)) #calling the method div() of the class math using the object calculator
```

```python
#with constructor example in data 
class Database:

    def __init__(self, server, database, username, password):
        self.server = server
        self.database = database
        self.username = username
        self.password = password

    def connect(self):
        print(f"Connecting to {self.server}")

    def read(self):
        print("Reading data")

    def write(self):
        print("Writing data")

db1 = Database("DEV","Snowflake","murali","password")
db2= Database("PROD","Snowflake","murali","password")

db1.connect()
db1.read()
db1.write()

db2.connect()
db2.read()
db2.write()

```

            Inheritance
Reuse the code
create child class from parent class

```python
#single level inheritance is the first type.
class dad1: #parent
    def house(self):
        print("my dad house")

class son(dad1): #child #single level
    def factory(self):
        print("my factory")

net=son()  ## net object is created only for son, but in method son is already linked to dad 
net.factory() # we can access son methods in net object 
net.house()  # And we can access dad methods in net object 
```

```python
#multilevel inheritance
class grandfather: #first parent #multilevel
    def car(self):
        print("my grandfather car")

class dad1(grandfather): #parent #multilevel
    def house(self):
        print("my dad house")

class son(dad1): #child
    def factory(self):
        print("my factory")

family=son() ##object for last class
family.car()
family.house()
family.factory()
```

```python
#hierarchy inheritance

class dad1: #parent
    def house(self):
        print("my dad house")

class son(dad1): #child #hierarchy
    def factory(self):
        print("my factory")

class son2(dad1): #child  #hierarchy
    def market(self):
        print("my market")

# for hierarchy we need to create seperate objects for each class which use same parent class

me=son()
me.house()
me.factory()
print ("seperater object for class which use same parent class")

brother=son2() ##create seperate object which use same parent class 
brother.house()
brother.market()
```

```python
#Mulitple inheritance

class dad1: #parent
    def house(self):
        print("my dad house")

class mom: #parent
    def shop(self):
        print("my mom shop")

# for mulitple inheritance we need link both parent class to access those different methods.
class daughter(dad1,mom): #child ##mulitple inheritance 
    def market(self):
        print("my market")

me=daughter()
me.house()
me.shop()
```

        Polymorphism

```python

#Polymorphism means the same method name can behave differently depending on the object that calls it.
class app1:
    def v1(self):
        print("one")
class app1_1(app1): 
    def v2(self):
        print("two")

    def v1(self):  ##overriding the app1 v1 logic 
        print("three")

live1=app1_1()
live1.v1()   ## V1 will fetched from last child not from parent if same method name becuase it will override the parent 
live1.v2()  
```
