        File handaling 

| Mode   | Description                                                   |
| ------ | ------------------------------------------------------------- |
| `'r'`  | Read-only (file must exist)                                   |
| `'w'`  | Write-only (overwrites existing file or creates a new file)   |
| `'a'`  | Append-only (adds data to the end of the file)                |
| `'r+'` | Read + Write (file must exist)                                |
| `'w+'` | Write + Read (overwrites existing file or creates a new file) |
| `'a+'` | Append + Read (creates the file if it doesn't exist)          |
| `'rb'` | Read binary                                                   |
| `'wb'` | Write binary                                                  |
| `'ab'` | Append binary                                                 |


```python
##creating new file

file=open("notes.txt","w")  ##open the file or create new file W will write the file
file.write("creating new file \n") ##write in the file
file.write("this is new file \n")
file.close() ##close the file 
```

```python
file =open("notes.txt","r")
content= file.read() ## we can store the data in sme varaibale 
print("the file data: \n", file.read()) ## if we multiple file read it will continue from where the above one left
print("Print from varaible: \n",content)

file1 =open("notes.txt","r")
content1=file1.readline() ## readline will check line by line in the file so memory will be fast
print("Print for readline: \n",content1)
file.close()

```

```python
##append
file =open("notes.txt","a")
file.write("\n add 3rd line")
file.close()
```

```python
##WITH 
##if we are usign with clause then no need to close the file everytime

with open ("notes.txt","r") as file:
    for line in file:    ## this part will call readline in backend ## for loop used to read the file line by line 
        print(line.strip())## strip will remove space in first and last
```

```python
##read the file line by line 
with open("notes.txt") as f:
    while True:
        line=f.readline()
        if not line:
            break
        if "error" in line :
            print("fount error:",line.strip())
```

```python
with open("notes.txt") as f:
    for _ in range(5): ##pagenation ##range is like filter only first 5 rows
        ## _ this is throuh awaya varaible no need of any 
        print(f.readline().strip())
```

```python
## load one file data to another file
import csv 

with open("input_file.csv","r") as in_file, open("output_file.csv","w") as out_file:
    for line in in_file:
        print(line.strip())
        out_file.write(line)
```

```python
## filter the value based column using library funation 

with open("input_file.csv","r") as in_file:
    reader=csv.DictReader(in_file) ## call this library to convert the data as dictonary and store the header
    
    for i in reader:
        print(i["age"])  ## based on column name 
```

```python
## filter the value based column using split or by using array method 
import csv 
with open("input_file.csv","r") as in_file:
    row1= in_file.readlines() ## this will store the row data to row1
    for i in row1[1:]: ## first row will be skipped 
        column=i.strip().split(",")
        print(column[2])
```
