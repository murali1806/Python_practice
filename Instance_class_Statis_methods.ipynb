        class method

class method is like call the method directly without object, directly calling class and once @class method is defined we can update any variable whoch defined in the class it will update the global defined varaible

```python
class employee:
    company ="tcs"
    @classmethod  
    def change_company(cls,name):  ##if class method is defined use cls inside method
        cls.company=name   ## this class method will update any varaible which defined inside class 

employee.change_company("Infosys")
print(employee.company)
```

        Static method

static method is like call the method directly without object, directly calling class and once @static method is defined we can use any variable only inside this method. it won't update the global defined varaible

```python
class employee:
    company ="tcs"
    @staticmethod  
    def change_company(name):  ##if statics method is defined no need to use self or cls
        company=name   ## this will update the value only inside this static method
        print("static method result inside",company) 

employee.change_company("Infosys")
print(employee.company)
```
