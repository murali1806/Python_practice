```python
##while loop
pin= input("enter pin")
correct_pin='1234'
i=0
print ("validating the pin")
while pin!=correct_pin:
    print("pin is incorrect")
    i+=1
    if(i==3):
        print("you have entered wrong pin 3 times")
        break
    pin= input("enter correct pin")

if pin==correct_pin:
    print("pin is correct")
```

```python
name=[1,5,2,3,4,6,7,8,9]
for i in name:
    if(name[i]==5): ##name[i] is the value of the list at index i
        break
print(i,name[i])

for i in range(len(name)):
    if(i==5):
        break ## break will exit the loop when i is 5
print(i,name[i])

for i in name:
    if(i==5):
        break
print(i,name[i])

continue ##continue will skip the current iteration and move to the next iteration of the loop
pass ##pass is a null statement in python. It is used as a placeholder for future code. It is used when a statement is required syntactically but you do not want any command or code to execute.
```

```python
def add(*args): ##*args accept any number of entries like list
    total=0
    for num in args: ##to read list we gave for loop 
        total+=num
    return total

print(add(1,3,2))
```

```python
## dictonary using kwargs with key and value sent to funcation 

def details(**kwargs):
    print("details")
    for key,value in kwargs.items(): ##key, value is syntax in for loo[]
        print(f"{key} :{value}")

details(name="murali",age=27,location="bangalore")
```
